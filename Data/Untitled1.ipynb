```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample Data
data = {
    'x': [1, 2, 3, 4, 5],
    'y': [2, 3, 5, 7, 11]
}
df = pd.DataFrame(data)

# Plotting the line graph
sns.lineplot(x='x', y='y', data=df)
plt.xlabel('X')
plt.ylabel('Y')
plt.savefig('line.png', dpi=300)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

def FUNC_Z(x, y):
    return 50 - (x ** 2 + y ** 2)

sns.set_style("whitegrid")
N = 50  # Adjusting the number of points for clearer edges

X_VAL = np.linspace(-5, 5, N)
Y_VAL = np.linspace(-5, 5, N)

X1, Y1 = np.meshgrid(X_VAL, Y_VAL)
Z1 = FUNC_Z(X1, Y1)

# Create the plot
fig = plt.figure()
axes = fig.add_subplot(111, projection='3d')

# Plot the surface with a gradient color and visible edges
surface = axes.plot_surface(X1, Y1, Z1, cmap='viridis', edgecolor='k', linewidth=0.2)

# Add a color bar which maps values to colors
fig.colorbar(surface, shrink=0.5, aspect=5)

# Show the plot
plt.savefig('3dmesh.png', dpi=300)
plt.show()
```

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import axes3d

def sample_spherical(npoints, ndim=3):
    vec = np.random.randn(ndim, npoints)
    vec /= np.linalg.norm(vec, axis=0)
    return vec

phi = np.linspace(0, np.pi, 20)
theta = np.linspace(0, 2 * np.pi, 40)
x = np.outer(np.sin(theta), np.cos(phi))
y = np.outer(np.sin(theta), np.sin(phi))
z = np.outer(np.cos(theta), np.ones_like(phi))

xi, yi, zi = sample_spherical(80)

fig, ax = plt.subplots(1, 1, subplot_kw={'projection':'3d', 'aspect':'auto'})
ax.plot_wireframe(x, y, z, color='k', rstride=1, cstride=1)
ax.scatter(xi, yi, zi, s=100, c='r', zorder=10)
plt.savefig('3dplanar.png', dpi=300)
```
